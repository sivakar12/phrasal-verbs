```python
import json
from bs4 import BeautifulSoup 
```

```python
with open('phrasal-verbs.xml') as f:
    html = f.read()
```

```python
soup = BeautifulSoup(html)
soup
```

```python
def get_example_with_blanks(example_tokens):
    if not example_tokens:
        return None
    def _replace(t):
        if str(t).startswith('<strong>'):
            return '__________'
        else:
            return t
    return ''.join([_replace(token) for token in example_tokens])
def get_example_answer(example_tokens):
    return '..'.join([token.text for token in example_tokens if str(token).startswith('<strong>')])
    return ''.join([_replace(token) for token in example_tokens])
def get_items_from_tr(tr):
    tds = tr.find_all('td')
    return {
        'verb': tds[0].text,
        'meaning': tds[1].text,
        #'example': tds[2].contents,
        'example_with_blanks': get_example_with_blanks(tds[2].contents),
        'example_answer': get_example_answer(tds[2].contents)
    }
items = [get_items_from_tr(item) for item in soup.find_all('tr')[1:]]
```

```python
json_dump = json.dumps(items)
```

```python
with open('phrasal_verbs_target.json', 'w') as out:
    out.write(json_dump)
```
