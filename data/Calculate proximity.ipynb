```python
import numpy as np
import pandas as pd
from nltk.tokenize import word_tokenize
import re
import spacy
import json
```

```python
df = pd.read_json('phrasal_verbs_target.json')
df.set_index('id', inplace=True)
df.head()
```

```python
df['combined'] = df['verb'] + ' ' + df['meaning'] + ' ' + df['example_with_blanks'] + ' '+ df['example_answer']
```

```python
df['combined'].head()
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp("Apple is looking at buying U.K. startup for $1 billion")
```

```python
similarities = np.zeros((len(df), len(df)))
```

```python
df.index.tolist()
```

```python
docs = {}
ids = [int(i) for i in df.index.tolist()]
for id in ids:
    docs[id] = nlp(df['verb'][id])
for i in ids:
    for j in ids:
        similarities[i][j] = docs[i].similarity(docs[j])
```

```python
similarities_df = pd.DataFrame(similarities, columns=df.index, index=df.index)
similarities_df
```

```python
similarities_df.describe()
```

```python
def get_closest_items(id):
    row = similarities_df[id]
    closest_ids = row.sort_values(ascending=False)[:11].index.tolist()
    return closest_ids
get_closest_items(2)
```

```python
with open('phrasal_verbs_target.json') as f:
    json_data = json.loads(f.read())
```

```python
for item in json_data:
    item['closest_items'] = [str(i) for i in get_closest_items(int(item['id']))]
```
